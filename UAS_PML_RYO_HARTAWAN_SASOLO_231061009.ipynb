```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

Kode ini bertujuan untuk membaca dataset dari file CSV dengan menggunakan pustaka pandas dan menampilkan kontennya.

Menampilkan isi DataFrame data.

```python
# Load the dataset
data = pd.read_csv('/content/used_car_data_new.csv', delimiter=';')
data
```

Kode ini digunakan untuk mengonversi data kategorikal menjadi numerik menggunakan LabelEncoder dari scikit-learn.

```python
# Encode categorical variables
label_encoders = {}
for column in ['id_merk', 'type', 'model', 'color', 'id_fuel_type']:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

Kode ini digunakan untuk memisahkan fitur (X) dan target (y) dalam dataset untuk keperluan model machine learning.

drop(): Fungsi pandas yang digunakan untuk menghapus kolom dari DataFrame.

columns=['id_ucd', 'mileage']: Menentukan kolom-kolom yang ingin dihapus dari DataFrame. Dalam hal ini, kolom yang dihapus adalah id_ucd dan mileage.

Hasil dari operasi ini adalah DataFrame baru yang tanpa kolom id_ucd dan mileage.

Menyimpan hasil DataFrame setelah kolom-kolom tersebut dihapus. Variabel X biasanya digunakan untuk fitur-fitur (independent variables) dalam model machine learning.

Kolom-kolom lainnya (selain id_ucd dan mileage) dianggap sebagai fitur atau input yang digunakan untuk memprediksi nilai target.

data['mileage']: Mengakses kolom mileage dalam DataFrame data.
y:
Menyimpan kolom mileage sebagai target variable atau dependent variable. Variabel ini adalah nilai yang ingin diprediksi oleh model machine learning berdasarkan fitur yang ada di X.

Dalam kasus ini, y berisi data tentang mileage atau jarak tempuh mobil yang digunakan sebagai target yang akan diprediksi.

```python
# Define features and target
X = data.drop(columns=['id_ucd', 'mileage'])
y = data['mileage']
```

Kode ini digunakan untuk normalisasi atau standardisasi fitur dalam dataset agar model machine learning dapat bekerja lebih baik dan lebih cepat.

StandardScaler adalah kelas dari pustaka scikit-learn yang digunakan untuk standardisasi data.

Fungsi utama dari StandardScaler adalah untuk mengubah fitur-fitur dalam dataset sehingga memiliki distribusi dengan rata-rata 0 dan simpangan baku 1. Ini berarti data akan memiliki skala yang sama.

Pentingnya Standardisasi:
Beberapa algoritma machine learning, seperti regresi linear, K-Nearest Neighbors (KNN), dan SVM, lebih sensitif terhadap perbedaan skala antar fitur.

Fitur dengan skala besar dapat mendominasi proses pelatihan dan mengakibatkan model tidak bekerja dengan optimal. Standardisasi mengatasi masalah ini.

fit_transform(X): Fungsi ini memiliki dua langkah:
fit(X): Menghitung statistik dari data, yaitu rata-rata dan simpangan baku (standard deviation) untuk setiap kolom fitur dalam X.

transform(X): Menggunakan nilai rata-rata dan simpangan baku yang dihitung untuk mengubah data X agar memiliki distribusi dengan rata-rata 0 dan simpangan baku 1.

Hasil dari fit_transform(X) adalah DataFrame baru (X_scaled) yang berisi data yang sudah distandardisasi.

```python
# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Kode ini digunakan untuk memisahkan dataset menjadi dua bagian: satu untuk pelatihan (training) dan satu lagi untuk pengujian (testing). Tujuan dari pemisahan ini adalah untuk menguji model pada data yang tidak terlihat sebelumnya (data testing) untuk mengevaluasi kinerja model secara objektif.

train_test_split digunakan untuk membagi data menjadi dua bagian: satu untuk pelatihan dan satu untuk pengujian.

Ini penting untuk menghindari overfitting dan untuk evaluasi objektif dari kinerja model.

test_size=0.2 menentukan 20% data untuk pengujian, dan random_state=42 memastikan hasil yang konsisten setiap kali kode dijalankan.

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Kode ini digunakan untuk membangun dan melatih model regresi linear menggunakan data pelatihan yang telah dipersiapkan sebelumnya.

LinearRegression() digunakan untuk membuat model regresi linear.

fit(X_train, y_train) melatih model dengan data pelatihan, menentukan hubungan terbaik antara fitur dan target.

Model yang sudah dilatih dapat digunakan untuk memprediksi nilai target pada data baru berdasarkan hubungan yang dipelajari.

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

Kode ini digunakan untuk memprediksi nilai target (output) berdasarkan data pengujian (features dari data yang belum dilihat sebelumnya) menggunakan model regresi linear yang sudah dilatih.

predict() adalah metode yang digunakan untuk membuat prediksi berdasarkan fitur input yang diberikan (dalam hal ini X_test).

Fungsi ini menggunakan model yang sudah dilatih sebelumnya (yang koefisiennya telah dihitung oleh metode fit()), dan memprediksi output (nilai target) untuk data yang tidak terlihat sebelumnya (dalam hal ini X_test).
X_test adalah data pengujian yang telah dipersiapkan sebelumnya, yang berisi fitur-fitur yang ingin kita prediksi.

Hasil dari prediksi ini adalah nilai-nilai yang diprediksi untuk target (yaitu, output yang model prediksikan berdasarkan fitur yang ada pada X_test).

predict(X_test) digunakan untuk memprediksi nilai target berdasarkan data pengujian X_test.

y_pred berisi nilai-nilai target yang diprediksi oleh model regresi linear.

Fungsi ini penting untuk mengevaluasi kemampuan model dalam memprediksi hasil pada data yang belum pernah dilihat sebelumnya, yang membantu dalam mengukur akurasi dan kinerja model.

```python
# Make predictions
y_pred = model.predict(X_test)
```

Kode ini digunakan untuk menilai kinerja model regresi dengan menghitung dua metrik evaluasi yang umum digunakan dalam regresi: Mean Squared Error (MSE) dan R-squared (R²).

**Tujuan dari Penggunaan MSE dan R²**

MSE (Mean Squared Error): Digunakan untuk menilai besarnya kesalahan model. Semakin kecil MSE, semakin baik model dalam memprediksi nilai target.

MSE berguna untuk memahami seberapa jauh prediksi model dari nilai sebenarnya dalam satuan kuadrat (sehingga lebih sensitif terhadap kesalahan besar).

R² (R-squared): Digunakan untuk menilai seberapa baik model menjelaskan variabilitas data target. Ini memberi gambaran tentang kekuatan hubungan antara fitur dan target.

R² yang lebih tinggi menunjukkan bahwa model berhasil menjelaskan banyak variasi dalam data target.

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

Kode ini digunakan untuk menampilkan koefisien regresi linear yang telah dipelajari oleh model setelah proses pelatihan. Koefisien ini menunjukkan hubungan antara fitur dan target dalam model regresi linear.

**Tujuan dan Hasil dari Kode Ini**

Tujuan: Untuk menunjukkan kontribusi setiap fitur terhadap hasil prediksi model. Koefisien yang lebih besar menunjukkan fitur yang lebih kuat pengaruhnya terhadap target, sedangkan koefisien yang lebih kecil atau mendekati nol menunjukkan fitur dengan pengaruh yang lebih lemah.

Hasil: Kode ini akan menampilkan daftar fitur dengan koefisien mereka yang diurutkan dari yang terbesar ke yang terkecil. Ini memberikan wawasan tentang fitur mana yang memiliki dampak paling besar dalam prediksi yang dihasilkan oleh model regresi linear.

Kode ini digunakan untuk menampilkan koefisien model regresi linear yang menunjukkan kontribusi setiap fitur terhadap target.

Dengan mengurutkan koefisien dari yang terbesar hingga terkecil, kita dapat menilai pengaruh setiap fitur terhadap prediksi model.

Ini berguna untuk memahami hubungan antara fitur dan target, serta menentukan fitur mana yang paling memengaruhi model.

```python
# Display model coefficients
coefficients = pd.Series(model.coef_, index=data.drop(columns=['id_ucd', 'mileage']).columns)
print("\nModel Coefficients:\n", coefficients.sort_values(ascending=False))
```

Kode ini digunakan untuk memvisualisasikan perbandingan antara nilai sebenarnya (actual values) dan nilai yang diprediksi (predicted values) dalam sebuah plot scatter plot, serta menunjukkan garis referensi yang menghubungkan titik-titik tersebut. Tujuan dari visualisasi ini adalah untuk memberikan gambaran seberapa baik model regresi dalam memprediksi nilai yang tepat.

```python
# Plot Actual vs Predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--r', linewidth=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Mileage')
plt.grid(True)
plt.show()
```

Kode ini digunakan untuk memvisualisasikan koefisien fitur (feature importance) yang dihasilkan oleh model regresi linier dalam bentuk bar chart horizontal. Bar chart ini membantu untuk memahami pengaruh relatif setiap fitur dalam memprediksi variabel target.

Fitur dengan koefisien terbesar (positif atau negatif) adalah fitur yang memiliki pengaruh paling kuat terhadap prediksi target.

Fitur dengan koefisien lebih kecil (lebih dekat dengan nol) memiliki pengaruh yang lebih kecil atau bahkan tidak berpengaruh signifikan terhadap prediksi.

Fitur dengan koefisien negatif berarti bahwa ketika nilai fitur tersebut meningkat, nilai prediksi target akan menurun, sementara fitur dengan koefisien positif berarti bahwa peningkatan nilai fitur akan menyebabkan peningkatan nilai target.

```python
# Plot feature importance
plt.figure(figsize=(10, 6))
coefficients.sort_values().plot(kind='barh', color='teal')
plt.title('Feature Importance (Linear Regression Coefficients)')
plt.xlabel('Coefficient Value')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```

Fungsi yang ada dalam kode tersebut digunakan untuk membangun dan melatih model Support Vector Regression (SVR) dengan menggunakan kernel Radial Basis Function (RBF).

Kode ini membangun model Support Vector Regression (SVR) dengan kernel RBF untuk memodelkan hubungan non-linear antara fitur input X_train dan nilai target y_train. Dengan menggunakan kernel RBF, model ini diharapkan dapat menangkap pola kompleks dalam data dan menghasilkan prediksi yang lebih akurat.

```python
# Create and train the model using Support Vector Regression (SVR)
model = SVR(kernel='rbf')
model.fit(X_train, y_train)
```

Kode yang diberikan digunakan untuk melakukan cross-validation pada model Support Vector Regression (SVR) dan menghitung rata-rata Mean Squared Error (MSE) berdasarkan hasil cross-validation. Cross-validation membantu mengevaluasi kinerja model dengan cara membaginya menjadi beberapa subset, melatih model pada beberapa bagian data, dan menguji kinerjanya pada bagian lainnya.

Proses ini memungkinkan Anda untuk mengevaluasi kinerja model Support Vector Regression (SVR) menggunakan cross-validation dengan 5 fold, dan menghitung rata-rata MSE yang menunjukkan tingkat kesalahan prediksi model. Hasil yang lebih rendah untuk MSE menunjukkan kinerja model yang lebih baik.

```python
# Perform cross-validation
cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
mean_cv_mse = -cv_scores.mean()
print(f"Mean Cross-Validated MSE: {mean_cv_mse}")
```

Fungsi predict() digunakan untuk menghasilkan prediksi berdasarkan model yang telah dilatih.

X_test adalah data input (fitur) yang digunakan untuk menguji model setelah model selesai dilatih pada data pelatihan (training data). Biasanya, X_test adalah bagian dari dataset yang tidak digunakan selama pelatihan model, sehingga memberikan gambaran tentang bagaimana model akan berperforma pada data yang tidak terlihat sebelumnya.

Fungsi ini akan mengembalikan nilai prediksi untuk setiap data input dalam X_test, berdasarkan pemahaman model terhadap hubungan antara fitur dan target yang telah dipelajari selama pelatihan.

Fungsi model.predict(X_test) memungkinkan kita untuk menghasilkan prediksi dari model yang telah dilatih pada data yang belum terlihat sebelumnya. Ini adalah langkah yang penting dalam evaluasi model, karena kita dapat membandingkan nilai prediksi (y_pred) dengan nilai aktual (y_test) untuk mengukur seberapa baik model dalam melakukan prediksi pada data yang tidak dikenal.

```python
# Make predictions
y_pred = model.predict(X_test)
```

Fungsi berikut digunakan untuk mengukur kinerja model dengan dua metrik evaluasi utama: Mean Squared Error (MSE) dan R-squared (R²), yang keduanya digunakan untuk menilai seberapa baik model dalam memprediksi nilai target berdasarkan data uji.

MSE memberikan gambaran kuantitatif tentang besarnya kesalahan prediksi model dalam satuan kuadrat. Metrik ini mengukur ketepatan model dengan menghitung rata-rata kesalahan kuadrat.

R² mengukur sejauh mana model dapat menjelaskan variabilitas data, dengan nilai yang lebih tinggi menunjukkan model yang lebih baik dalam mencocokkan data.

Kedua metrik ini memberikan pandangan yang saling melengkapi tentang kinerja model, dengan MSE mengukur besarnya kesalahan dan R² mengukur seberapa baik model menjelaskan variasi dalam data.

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

k = 5:

Variabel k diatur ke nilai 5, yang menentukan jumlah tetangga terdekat yang digunakan oleh algoritma K-Nearest Neighbors. Dalam hal ini, model akan menggunakan 5 tetangga terdekat untuk membuat prediksi nilai target.

Dalam KNN, nilai k adalah parameter yang penting untuk kontrol bagaimana model membuat prediksi berdasarkan data yang ada. Nilai k yang lebih kecil cenderung lebih sensitif terhadap noise, sementara nilai k yang lebih besar dapat menghasilkan prediksi yang lebih stabil tetapi mungkin kurang akurat pada data yang kompleks.

model = KNeighborsRegressor(n_neighbors=k):

Baris ini membuat sebuah model K-Nearest Neighbors untuk regresi menggunakan kelas KNeighborsRegressor dari scikit-learn.

n_neighbors=k adalah parameter yang digunakan untuk menentukan jumlah tetangga yang dipertimbangkan dalam proses prediksi. Dalam hal ini, k=5 berarti model akan mempertimbangkan 5 tetangga terdekat (dalam ruang fitur) untuk menentukan nilai target yang diprediksi.

KNN adalah metode berbasis jarak yang digunakan untuk memprediksi nilai target berdasarkan nilai terdekat dari data pelatihan. Dalam regresi, KNN akan mengambil rata-rata nilai target dari k tetangga terdekat untuk membuat prediksi.

model.fit(X_train, y_train):

Fungsi fit() digunakan untuk melatih model menggunakan data pelatihan.

X_train adalah data fitur (independen) yang digunakan untuk melatih model. Ini adalah subset dari data yang sudah diproses dan memiliki nilai yang ingin diprediksi.

y_train adalah nilai target (tergantung) yang sebenarnya untuk data pelatihan. Ini adalah nilai yang ingin diprediksi oleh model.

Dengan menggunakan data pelatihan ini, model KNN akan menyimpan informasi mengenai data untuk dapat menghitung jarak antar titik dan memprediksi nilai berdasarkan tetangga terdekat saat diuji nanti.

```python
# Create and train the model using K-Nearest Neighbors (KNN)
k = 5  # Number of neighbors
model = KNeighborsRegressor(n_neighbors=k)
model.fit(X_train, y_train)
```

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error'):

cross_val_score() adalah fungsi dari scikit-learn yang digunakan untuk melakukan cross-validation pada model.

Cross-validation adalah teknik untuk mengevaluasi model dengan membaginya menjadi beberapa subset (fold), melatihnya pada subset tertentu, dan menguji pada subset lainnya. Ini membantu untuk mendapatkan estimasi kinerja model yang lebih stabil dan menghindari overfitting.

Parameter:

model: Model yang akan dievaluasi. Dalam hal ini, model yang digunakan adalah KNeighborsRegressor yang telah dilatih sebelumnya.

X_scaled: Data fitur yang sudah di-scaling (misalnya menggunakan StandardScaler). Ini adalah data yang digunakan untuk melatih model.

y: Data target yang ingin diprediksi oleh model.

cv=5: Menentukan jumlah folds untuk cross-validation.

Dengan cv=5, data akan dibagi menjadi 5 bagian, dan model akan dilatih dan diuji sebanyak 5 kali, masing-masing kali dengan data pelatihan dan pengujian yang berbeda.

scoring='neg_mean_squared_error': Menentukan metrik evaluasi yang digunakan untuk menilai kinerja model.

Dalam hal ini, digunakan Mean Squared Error (MSE), tetapi karena fungsi cross_val_score mengembalikan nilai positif untuk skor yang lebih baik, maka nilai MSE dihitung dalam bentuk negatif (neg_mean_squared_error).

 Ini memungkinkan kita untuk mendapatkan nilai negatif yang bisa diubah menjadi nilai positif untuk perhitungan rata-rata MSE.
Fungsi ini mengembalikan array nilai skor MSE negatif untuk setiap fold dari cross-validation.

mean_cv_mse = -cv_scores.mean():

cv_scores.mean() menghitung rata-rata dari nilai MSE negatif yang diperoleh dari setiap fold.

Mengingat nilai yang dikembalikan oleh cross_val_score adalah negatif (karena scoring menggunakan 'neg_mean_squared_error'), kita perlu mengubah tanda menjadi positif menggunakan tanda negatif (-cv_scores.

mean()). Ini memberikan nilai rata-rata MSE untuk cross-validation.
print(f"Mean Cross-Validated MSE: {mean_cv_mse}"):

Baris ini mencetak nilai rata-rata MSE yang dihitung dari cross-validation.

Mean Cross-Validated MSE menggambarkan seberapa baik model KNN dalam memprediksi target pada data uji yang belum terlihat selama pelatihan, dengan mempertimbangkan fluktuasi hasil yang dihasilkan oleh pembagian data menjadi beberapa fold.

```python
# Perform cross-validation
cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
mean_cv_mse = -cv_scores.mean()
print(f"Mean Cross-Validated MSE: {mean_cv_mse}")
```

model.predict(X_test):

model: Ini adalah model yang telah dilatih sebelumnya, seperti KNeighborsRegressor, LinearRegression, atau model lainnya. Model ini telah dipelajari dengan data pelatihan (train data), dan sekarang digunakan untuk membuat prediksi pada data yang belum dilihat oleh model, yaitu data uji (X_test).

X_test: Ini adalah data uji (test data) yang berisi fitur-fitur (independent variables) yang akan digunakan untuk membuat prediksi. Data ini biasanya merupakan subset dari dataset yang lebih besar dan dipisahkan dari data pelatihan untuk mengevaluasi kinerja model. X_test tidak termasuk target (dependent variable atau label), yang hanya digunakan untuk evaluasi.

Proses Prediksi:

Ketika kita memanggil predict() pada model, model akan menggunakan parameter yang telah dipelajari selama pelatihan untuk memprediksi target (nilai yang ingin diprediksi) berdasarkan fitur-fitur yang ada pada X_test.

Misalnya, jika kita menggunakan regresi (seperti Linear Regression atau KNN), model akan menghasilkan nilai numerik yang diprediksi berdasarkan hubungan yang dipelajari antara fitur (X_test) dan target yang sebenarnya (pada data pelatihan).

Output y_pred:

y_pred adalah hasil prediksi dari model untuk setiap entri dalam X_test. Ini berisi nilai-nilai yang diprediksi oleh model berdasarkan fitur yang ada di X_test.

Dalam regresi, y_pred akan berisi nilai-nilai numerik yang diperkirakan oleh model untuk variabel target yang ingin diprediksi.

```python
# Make predictions
y_pred = model.predict(X_test)
```

Kode ini digunakan untuk menilai kinerja model regresi dengan menghitung dua metrik evaluasi yang umum digunakan dalam regresi: Mean Squared Error (MSE) dan R-squared (R²).

**Tujuan dari Penggunaan MSE dan R²**

MSE (Mean Squared Error): Digunakan untuk menilai besarnya kesalahan model. Semakin kecil MSE, semakin baik model dalam memprediksi nilai target.

MSE berguna untuk memahami seberapa jauh prediksi model dari nilai sebenarnya dalam satuan kuadrat (sehingga lebih sensitif terhadap kesalahan besar).

R² (R-squared): Digunakan untuk menilai seberapa baik model menjelaskan variabilitas data target. Ini memberi gambaran tentang kekuatan hubungan antara fitur dan target.

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
